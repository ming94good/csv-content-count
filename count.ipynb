```python
import csv
import pandas as pd

# 開啟 CSV 檔案
def caculate(table,column):
    with open(table, newline='') as csvfile:
      content = csv.reader(csvfile)
      # read column number (10-13)
      if(column is 'all'):
        included_cols = [10,11,12,13]
      elif(10 <= int(column) and int(column) <= 13):
        included_cols = [int(column)] 
      else:
        return 'bad input (Type in: 10-13 or all)'
      total_songs = []
      for rows in content:
        song = list(rows[i] for i in included_cols)
        total_songs.append(song)
      total_songs.pop(0)
      song_list = []
      for songs in total_songs:
        for song_content in songs:
            song = song_content.split('\n')
            for s in song:
                if(s is not ' ' and s is not ''):
                  song_list.append(s)
    # Pandas value count
    a = pd.DataFrame(song_list)
    pd.set_option("display.max_rows", None, "display.max_columns", None)
    result = a.apply(pd.value_counts)
    print(result)
```

### 第一首

```python
caculate('table.csv','10')
```

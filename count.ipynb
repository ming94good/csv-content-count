```python
import csv
import pandas as pd

# 開啟 CSV 檔案
with open('Desktop/table.csv', newline='') as csvfile:

  # 讀取 CSV 檔案內容
  content = csv.reader(csvfile)

  # 以迴圈輸出每一列
  #for row in rows:
    #print(row)
  included_cols = [10,11,12] 
  total_songs = []
  for rows in content:
    song = list(rows[i] for i in included_cols)
    total_songs.append(song)
```

```python
song_list = []
for songs in total_songs:
    for song in songs:
        song_list.append(song)
```

```python
# remove noise
song_list.pop(0)
song_list.pop(0)
song_list.pop(0)
```

```python
# Pandas value count
a = pd.DataFrame(song_list)
pd.set_option("display.max_rows", None, "display.max_columns", None)
result = a.apply(pd.value_counts)
print(result)
```

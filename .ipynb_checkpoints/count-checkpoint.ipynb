```python
import csv
import pandas as pd

# 開啟 CSV 檔案
with open('table.csv', newline='') as csvfile:
  content = csv.reader(csvfile)
  # read column number (10-13)
#   included_cols = [10,11,12] 
#   total_songs = []
#   for rows in content:
#     song = list(rows[i] for i in included_cols)
#     total_songs.append(song)
#   total_songs.pop(0)
#   song_list = []
#   for songs in total_songs:
#     for song in songs:
#         song_list.append(song)
        
  included_cols = [13] 
  total_songs = []
  for rows in content:
    song = list(rows[i] for i in included_cols)
    total_songs.append(song)
  total_songs.pop(0)
  song_list = []
  for song in total_songs:
    for s in song:
        song_list.append(s)
```

```python
for song in total_songs:
    for s in song:
        print(s)
```

```python
# Pandas value count
a = pd.DataFrame(song_list)
pd.set_option("display.max_rows", None, "display.max_columns", None)
result = a.apply(pd.value_counts)
# print(result)
```
